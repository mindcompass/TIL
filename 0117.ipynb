```python
import re
text = '010-8478-8181'
regex = '(\d{3})-\d{3,4}-(\d{4})'
pattern = re.compile(regex)
change = '\g<1>-****-\g<2>'
res = pattern.sub(change, text)
print(res)
```

```python
import re
user_list = ['ggoreb@naver.com', 'seorab@gmail.com', 'human@ggoreb.com']
regex = '(\w+@)\w+([.]com)'
pattern = re.compile(regex)
change = '\g<1>*****\g<2>'
for user in user_list:
    res = pattern.sub(change, user)
    print(res)
```

```python
import re
regex = '[0-9a-zA-Z_+.-]+@[a-zA-Z]+[.]\d+'
pattern = re.compile(regex)
emails = ['python@mail.ggoreb.com', 'python+kr@ggoreb.com','python-10@ggoreb.co.kr', 'python_10@ggoreb.info', 'python.10@exam-ggoreb.com',
 '@ggoreb.com', 'python@ggoreb', 'python@ggoreb-com']
for email in emails:
    res = pattern.search(email)
    if res:
        print(res.group())
```
